```python
import numpy as np
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
import warnings 
warnings.filterwarnings('ignore')
```

<h2>Load the dataset with the correct encoding</h2>

```python
df = pd.read_csv('dataset\\Global_terrorist.csv', encoding='ISO-8859-1')
```

```python
df.head(5)
```

<h2>Filtering the data.</h2>

```python
filter1 = df[df['region_txt'] == "Middle East & North Africa"]
filter2 = filter1[filter1['iyear'] >= 2000]
filter3 = filter2[filter2['attacktype1_txt'] >= 'Bombing/Explosion']
filter4 = filter3[filter3['nkill'] >= 10]
```

<h2>Data aggregation.</h2>

```python
spatial_data1 = filter4.groupby(['longitude', 'latitude']).size().reset_index(name='terror_count')
spatial_data1['normalized_weight'] = (spatial_data1['terror_count'] - spatial_data1['terror_count'].min()) /\
                                    (spatial_data1['terror_count'].max() - spatial_data1['terror_count'].min())
```

```python
spatial_data1
```

<h2>Preparing heatmap and the data.</h2>

```python
import folium
from folium.plugins import HeatMap
```

```python
spatial_list1 = spatial_data1[['longitude', 'latitude', 'normalized_weight']].values.tolist()
base_map = folium.Map(location=[30, 40], zoom_start=4)
HeatMap(spatial_list1).add_to(base_map)
base_map.save('myGeoSpatial.html')
base_map
```

<hr>

<h1>Regional Analysis</h1>

```python
region_fil1 = df[(df['region_txt'] == 'South Asia') & (df['iyear'] >= 2000) & (df['nkill'] >= 10)]

regional_spatial = region_fil1.groupby(['longitude', 'latitude']).size().reset_index(name='terror_count')
regional_spatial['normalized_weight'] = (regional_spatial['terror_count'] - regional_spatial['terror_count'].min()) /\
                                    (regional_spatial['terror_count'].max() - regional_spatial['terror_count'].min())


regional_spatial_list = regional_spatial[['longitude', 'latitude', 'normalized_weight']].values.tolist()
regional_base_map = folium.Map(location=[30, 40], zoom_start=2)
HeatMap(regional_spatial_list).add_to(regional_base_map)
regional_base_map.save('regional.html')
regional_base_map
```

```python
region_fil1['country_txt'].value_counts()
```

<h3>Insight #1 : Base on the analysis the data provided that in the south asia where kills reach 10 and above Afghanistan got the highest number of incident.</h3>

<h1>Crime Type Analysis</h1>

```python
crime_fl1 = df[(df['alternative_txt'] == 'Insurgency/Guerilla Action') & (df['iyear'] >= 2000) & (df['nkill'] >= 10)]

crime_spatial = crime_fl1.groupby(['longitude', 'latitude']).size().reset_index(name='terror_count')
crime_spatial['normalized_weight'] = (crime_spatial['terror_count'] - crime_spatial['terror_count'].min()) /\
                                    (crime_spatial['terror_count'].max() - crime_spatial['terror_count'].min())


crime_spatial_list = crime_spatial[['longitude', 'latitude', 'normalized_weight']].values.tolist()
crime_base_map = folium.Map(location=[30, 40], zoom_start=4)
HeatMap(crime_spatial_list).add_to(crime_base_map)
crime_base_map.save('crime.html')
crime_base_map
```

<h3>Insight #2 : Base on the analysis the data provided that in the crime type of "Insurgency/Guerilla Action" where kills reach 10 and above is mainly concentrated in the Middle East & North Africa region and Iraq got the highest number of incident record.</h3>

<h1>Casualty Analysis</h1>

```python
casualty_fl1 = df[
    (df['iyear'] >= 2000) & 
    (df['nkill'] >= 20) & 
    (df['region_txt'].isin(['Middle East & North Africa', 'Sub-Saharan Africa', 'South Asia']))]

casualty_spatial = casualty_fl1.groupby(['longitude', 'latitude']).size().reset_index(name='terror_count')
casualty_spatial['normalized_weight'] = (casualty_spatial['terror_count'] - casualty_spatial['terror_count'].min()) /\
                                    (casualty_spatial['terror_count'].max() - casualty_spatial['terror_count'].min())

casualty_spatial_list = casualty_spatial[['longitude', 'latitude', 'normalized_weight']].values.tolist()
casualty_base_map = folium.Map(location=[30, 40], zoom_start=2)
HeatMap(casualty_spatial_list).add_to(casualty_base_map)
casualty_base_map.save('casualty.html')
casualty_base_map
```

<h3>Insight #3 : Base on the analysis the data provided that in the where kills reach 20 kills casualties and above is mainly concentrated in the Middle East & North Africa region 1066 cases followed by and Sub-Saharan Africa 667 cases, and South Asia 530 cases.</h3>
